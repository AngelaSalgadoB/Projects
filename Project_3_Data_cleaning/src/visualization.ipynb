```python
import numpy as np

average_salary_general=pd.read_sql_query(sql=queries_salaries.male_female_average_salary,con=con)
average_salary_general.columns = ['company','average_general']
print(average_salary_general.columns)

average_salary_male=pd.read_sql_query(sql=queries_salaries.gender_male_best_salaries_in_company,con=con)
average_salary_male.columns=['company','gender', 'average_male']
print(average_salary_male.columns)


average_salary_female=pd.read_sql_query(sql=queries_salaries.gender_female_best_salaries_in_company,con=con)
average_salary_female.columns=['company','gender', 'average_female']
print(average_salary_female.columns)

test = average_salary_general.merge(average_salary_male, how = "inner", on = "company")
test2 = test.merge(average_salary_female, how = "inner", on = "company")
print(test2.info())
print(test2.head())

test2 = test2.nlargest(40,"average_general")

sns.scatterplot(data=test2, x="average_general", y="average_male", size="average_female",sizes=(20,2000), legend="brief", hue='average_female')
# plt.xticks(np.arange(min(test2.average_general), max(test2.average_general)+1, 10000))
plt.xlim(170000,300000)
plt.ylim(170000,300000)
plt.subplots_adjust(wspace=10, bottom=0.3)


plt.show()
```
